```python
import torch
import math
from Cube import CubeFile, read_charges_refined, read_cube
import os
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from numba import jit
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

import pickle
import pandas as pd
```

```python
data = []
for i in range(1,8):
    from_pickle = "butadiene_{}.p".format(i)
    x = pickle.load(open(from_pickle, "rb")).x
    x_np = np.zeros((3,13))
    for i in range(len(x)):
        x_np[i % 3, i // 3] = x[i]
    data.append(x_np)

```

```python
distances = []
for i in range(6):
    dif = data[i] - data[i+1]
    distance = np.sqrt(np.square(dif.sum(axis=0)))
    distances.append(distance)

np.vstack(distances).T

fig, axs = plt.subplots(13, sharex=True)

i = 0
for x in np.vstack(distances).T:
    c = "k"
    axs[i].plot(np.arange(1, 7), x, c=c)
    #axs[i].scatter(np.arange(1, 7), x, c=c)
    axs[i].set_ylim(0, 6)
    axs[i].grid(axis="both", which="both")
    axs[i].xaxis.set_minor_locator(MultipleLocator(1))
    axs[i].xaxis.set_major_locator(MultipleLocator(1))
    axs[i].yaxis.set_minor_locator(MultipleLocator(1))
    axs[i].yaxis.set_major_locator(MultipleLocator(5))
    axs[i].text(1, 4, "c{}".format(i+1))
#     axs[i].axhline(0, c="k")
    i += 1
    
for ax in axs:
    ax.label_outer()

fig.text(0.04, 0.5, 'RMSD(N, N+1)', va='center', rotation='vertical', size=14)
fig.text(0.5, 0.08, 'Scan point N', ha='center', size=14)

fig.set_size_inches(6, 10)
```
